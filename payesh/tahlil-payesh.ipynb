```python
import pandas as pd
import numpy as np

# 1. Load the data with multi-level header
try:
    df = pd.read_excel("1401.xlsx", header=[0, 1])
except UnicodeDecodeError:
    df = pd.read_excel("1401.xlsx", header=[0, 1])

# 2. Clean Column Names
new_cols = []
for col_l1, col_l2 in df.columns:
    if 'Unnamed:' in col_l2:
         new_cols.append(col_l1.strip())
    else:
        new_cols.append(f"{col_l1.strip()}_{col_l2.strip()}")
df.columns = new_cols
```

```python

# 3. Identify ID Columns
id_vars = [
    'نام مشترک', 'کد اشتراک', 'نوع پروانه', 'استان', 'شهرستان', 'امور آب',
    'محدوده مطالعاتی', 'سریال کنتور', 'سایز کنتور', 'تاریخ نصب',
    'زمان آخرین اتصال', 'قطع کن', 'مصرف بازه (m³)', 'ساعت کارکرد بازه (h)',
    'میانگین دبی بازه l/s'
]
id_vars = [col for col in id_vars if col in df.columns]

# Identify value columns
value_vars = [col for col in df.columns if col not in id_vars]

# 4. Melt/Unpivot the DataFrame
df_melted = pd.melt(df,
                    id_vars=id_vars,
                    value_vars=value_vars,
                    var_name='Metric_Month_Raw',
                    value_name='Value')

# 5. Split Metric_Month_Raw into separate Month and Metric columns
df_melted[['Month', 'Metric']] = df_melted['Metric_Month_Raw'].str.split('_', n=1, expand=True)
df_melted = df_melted.drop('Metric_Month_Raw', axis=1)

# Ensure 'Value' is numeric (important for later use)
df_melted['Value'] = pd.to_numeric(df_melted['Value'], errors='coerce')

# print("\nHead of the Melted DataFrame (Recommended Format):")
# print(df_melted.head(10))
# print("\nInfo of the Melted DataFrame:")
# df_melted.info()
df_melted.drop(['Metric','Value'],axis=1,inplace=True)

# Save the melted result
output_filename = "temp1401.csv"
df_melted.to_csv(output_filename, index=False, encoding='utf-8-sig')
# print(f"\nSignificantly more readable melted data saved to {output_filename}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the Excel dataset
file_path = '1401.xlsx'
df = pd.read_excel(file_path, engine='openpyxl')
```

```python

# Cleaning the data: Replacing empty or invalid values with NaN and dropping rows with critical missing data
df.replace('', np.nan, inplace=True)
df.dropna(subset=['نام مشترک', 'کد اشتراک'], inplace=True)  # Dropping rows missing customer name or subscription code
```

```python

df.rename(columns=column_translations, inplace=True)
```

```python
# Converting numeric columns to appropriate types, handling invalid values
numeric_cols = ['Consumption in Period (m³)', 'Operating Hours in Period (h)', 'Average Flow Rate in Period (l/s)']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
import pandas as pd
import numpy as np
# --- 1. Header Renaming ---
# Define metadata headers
metadata_headers = [
    'UserName', 'UserID', 'UsageType', 'Province', 'City',
    'EmptyCol5', 'EmptyCol6', 'EmptyCol7', 'MeterID', 'CapacityOrPipeSize',
    'StartDate', 'LastReadingDateTime', 'StatusFlag', 'TotalValue1', 'TotalValue2'
]

# Define base names for the repeating monthly columns
monthly_base_headers = [
    'AvgFlow_LPS', 'NegFlowCount', 'NegFlowPercent', 'OperatingHours', 'Consumption_M3',
    'DataCountActual', 'DataCountExpected', 'DataAvailabilityPercent'
]

# Generate the full list of column names for the initial read
new_column_names = metadata_headers[:]  # Start with metadata headers
for i in range(1, 13):  # 12 months
    for base_header in monthly_base_headers:
        new_column_names.append(f"{base_header}_M{i}")

# --- Load the data ---
excel_file_path = "1401.xlsx" # Replace with your actual file path

try:
    # Tell pandas there is NO header row to read automatically (header=None)
    # Skip the first 2 rows which contain the original multi-level header (skiprows=2)
    # Apply your predefined list of names (names=new_column_names)
    df = pd.read_excel(
        excel_file_path,
        header=None,         # Don't parse any header from the file
        skiprows=2,          # Skip the actual header rows (row 0 and 1)
        names=new_column_names # Apply your custom names
    )
    print(f"Successfully loaded data from {excel_file_path}")
    print(f"Initial shape: {df.shape}")

except FileNotFoundError:
    print(f"Error: File not found at {excel_file_path}")
    exit() # Stop script if file not found
except Exception as e:
    print(f"An error occurred during file reading: {e}")
    exit() # Stop script on other reading errors

# --- Convert relevant columns to numeric ---
# Identify columns that should be numeric (all monthly metrics + Capacity + TotalValues)
numeric_cols = ['CapacityOrPipeSize', 'TotalValue1', 'TotalValue2']
for i in range(1, 13):
    for base_header in monthly_base_headers:
         # Always try to convert monthly metrics to numeric, coerce errors
         numeric_cols.append(f"{base_header}_M{i}")

# Use a set to avoid duplicates and then convert back to list
numeric_cols = list(set(numeric_cols))

print("\n--- Converting columns to numeric ---")
for col in numeric_cols:
    if col in df.columns: # Check if column exists
        # print(f"Converting {col}...") # Uncomment for debugging
        df[col] = pd.to_numeric(df[col], errors='coerce') # errors='coerce' turns non-numeric things into NaN
    # else: # Uncomment for debugging
        # print(f"Warning: Column {col} not found for numeric conversion.")

print("--- Dataframe head after renaming and type conversion ---")
print(df.head())
print("\n--- Dataframe info after renaming and type conversion ---")
df.info()

# --- 3. Investigate Anomalies (using the wide format 'df') ---
# Anomaly detection is often easier before reshaping

print("\n" + "="*50)
print("--- Anomaly Investigation (on Wide Data) ---")
print("="*50 + "\n")

# Anomaly 1: Negative Consumption
consumption_cols = [f'Consumption_M3_M{i}' for i in range(1, 13) if f'Consumption_M3_M{i}' in df.columns]
if consumption_cols: # Check if any consumption columns exist
    negative_consumption = df[(df[consumption_cols] < 0).any(axis=1)]
    print(f"--- Rows with Negative Monthly Consumption ({len(negative_consumption)}) ---")
    if not negative_consumption.empty:
        # Select only relevant columns to display
        display_cols_neg_consump = ['UserName', 'UserID'] + [col for col in consumption_cols if (negative_consumption[col] < 0).any()]
        print(negative_consumption[display_cols_neg_consump])
    else:
        print("No rows found with negative monthly consumption.")
else:
    print("Skipping negative consumption check - no consumption columns found.")

print("\n" + "-"*50 + "\n")


# Anomaly 2: Presence of Negative Flow Percentage
neg_flow_perc_cols = [f'NegFlowPercent_M{i}' for i in range(1, 13) if f'NegFlowPercent_M{i}' in df.columns]
if neg_flow_perc_cols: # Check if any neg flow columns exist
    # Check for any value greater than 0 in the negative flow percentage columns
    has_negative_flow_perc = df[(df[neg_flow_perc_cols] > 0).any(axis=1)]
    print(f"--- Rows with Indicated Negative Flow Percentage > 0% ({len(has_negative_flow_perc)}) ---")
    if not has_negative_flow_perc.empty:
        # Select relevant columns to display
        display_cols_neg_perc = ['UserName', 'UserID'] + [col for col in neg_flow_perc_cols if (has_negative_flow_perc[col] > 0).any()]
        print(has_negative_flow_perc[display_cols_neg_perc])
    else:
        print("No rows found with negative flow percentage > 0%.")
else:
    print("Skipping negative flow percentage check - no relevant columns found.")

print("\n" + "-"*50 + "\n")

# Anomaly 3: Zero or Near-Zero Consumption with High Data Availability
# Calculate total consumption (handle potential NaNs introduced by 'coerce')
if consumption_cols:
    df['TotalConsumption'] = df[consumption_cols].sum(axis=1, skipna=True) # skipna=True sums available data

    # Calculate average availability
    availability_cols = [f'DataAvailabilityPercent_M{i}' for i in range(1, 13) if f'DataAvailabilityPercent_M{i}' in df.columns]
    if availability_cols:
        df['AverageAvailability'] = df[availability_cols].mean(axis=1, skipna=True)

        # Define thresholds
        consumption_threshold = 0.1 # Allow for very small positive values or floating point inaccuracies
        availability_threshold = 90 # e.g., 90% average availability

        zero_consump_high_avail = df[
            (df['TotalConsumption'].abs() < consumption_threshold) &
            (df['AverageAvailability'] >= availability_threshold)
        ]

        print(f"--- Rows with Near-Zero Total Consumption (<{consumption_threshold}) AND High Avg Availability (>= {availability_threshold}%) ({len(zero_consump_high_avail)}) ---")
        if not zero_consump_high_avail.empty:
            display_cols_zero = ['UserName', 'UserID', 'UsageType', 'TotalConsumption', 'AverageAvailability']
            print(zero_consump_high_avail[display_cols_zero])
        else:
            print("No rows found matching the zero consumption / high availability criteria.")

        # Clean up added columns
        df = df.drop(columns=['TotalConsumption', 'AverageAvailability'], errors='ignore')
    else:
        print("Skipping zero consumption check - no availability columns found.")
else:
    print("Skipping zero consumption check - no consumption columns found.")

print("\n" + "="*50)
print("--- End of Anomaly Investigation ---")
print("="*50)


# --- START OF Reshape Wide to Long SECTION ---

print("\n" + "="*50)
print("--- Reshaping Data from Wide to Long ---")
print("="*50 + "\n")

# 1. Define the list of columns that identify a unique user/row (NOT the monthly data)
# These are your original metadata_headers
id_vars = metadata_headers

# 2. Define the base names (stubs) of the columns you want to stack.
# This is your original monthly_base_headers list.
stub_names = monthly_base_headers

# 3. Reset the index and add it to id_vars for unique row identification during melt
df_reset = df.reset_index()
id_vars_with_index = ['index'] + id_vars # Add the new 'index' column to id_vars

# 4. Perform the wide_to_long transformation
df_long = pd.DataFrame() # Initialize empty DataFrame for safety
try:
    # Ensure stub_names actually exist in the DataFrame columns before melting
    # We check based on the existence of the '_M1' version of the stub
    existing_stubs = [stub for stub in stub_names if f"{stub}_M1" in df_reset.columns]
    if not existing_stubs:
         print("Error: None of the specified monthly base headers (stubs) were found in the columns.")
         print("Cannot perform wide_to_long. Check 'monthly_base_headers' and column names in the DataFrame.")
    else:
        if len(existing_stubs) < len(stub_names):
            print("Warning: Some base metric headers were not found in the DataFrame columns and will be skipped.")
            missing_stubs = set(stub_names) - set(existing_stubs)
            print(f"Missing stubs (or their _M1 version): {list(missing_stubs)}")
            print(f"Using stubs found: {existing_stubs}")

        df_long = pd.wide_to_long(
            df_reset,
            stubnames=existing_stubs,   # Use only the stubs that actually exist
            i=id_vars_with_index,       # Columns to keep as identifiers
            j='MonthNumber',            # Name of the new column indicating the month (1-12)
            sep='_M',                   # Separator between the stub name and the month number
            suffix='\\d+'               # Pattern for the suffix (one or more digits)
        )

        # Reset the index to turn the MultiIndex (id_vars + MonthNumber) back into columns
        df_long = df_long.reset_index()

        # Optional: Remove the temporary 'index' column if you don't need it
        if 'index' in df_long.columns:
            df_long = df_long.drop(columns=['index'])

        # print("--- Dataframe head after reshaping to long format ---")
        # print(df_long.head(15)) # Print more rows to see different months for the first user
        # print("\n--- Dataframe info after reshaping ---")
        # df_long.info()
        # print(f"\nTotal rows in long format: {len(df_long)}")
        # print(f"Expected rows (approx): {len(df)} * 12 = {len(df)*12}")

except ValueError as ve:
     print(f"\nValueError during wide_to_long: {ve}")
     print("This often happens if suffix pattern doesn't match (e.g., '_M1_extra') or columns are missing.")
except Exception as e:
    print(f"\nAn unexpected error occurred during wide_to_long transformation: {e}")
    print("Please check your 'metadata_headers' and 'monthly_base_headers' lists match the DataFrame columns and structure.")
    # df_long remains empty or partially formed


# --- Example Analysis on df_long ---
if not df_long.empty:
    # print("\n" + "="*50)
    # print("--- Example Analysis on Long Data: Avg Consumption by Type/Month ---")
    # print("="*50 + "\n")

    # Fill potential NaNs in UsageType for grouping
    if 'UsageType' in df_long.columns:
        df_long['UsageType'] = df_long['UsageType'].fillna('Unknown')

        if 'Consumption_M3' in df_long.columns:
            avg_consumption_by_type_month = df_long.groupby(['UsageType', 'MonthNumber'])['Consumption_M3'].mean().reset_index()

            # Pivot for better readability (optional)
            try:
                avg_consumption_pivot = avg_consumption_by_type_month.pivot(
                    index='UsageType',
                    columns='MonthNumber',
                    values='Consumption_M3'
                )
                #print("--- Average Monthly Consumption (m³) by Usage Type ---")
                #print(avg_consumption_pivot)
            except Exception as pivot_e:
                #print(f"Could not pivot results: {pivot_e}")
                #print(avg_consumption_by_type_month) # Print the unpivoted results instead
                pass            
        else:
            pass
            #print("Skipping consumption analysis - 'Consumption_M3' column not found in long format.")

        # You can also easily analyze trends for specific users:
        if not df_long.empty and 'UserName' in df_long.columns:
            user_example = df_long['UserName'].iloc[0] # Get the first username
            # print(f"\n--- Monthly Data for User: {user_example} ---")
            display_cols_user = ['MonthNumber'] + [col for col in ['AvgFlow_LPS', 'OperatingHours', 'Consumption_M3', 'DataAvailabilityPercent'] if col in df_long.columns]
            # print(df_long[df_long['UserName'] == user_example][display_cols_user])
        else:
            print("Could not display user example - DataFrame is empty or 'UserName' column missing.")
    else:
        print("Skipping UsageType analysis - 'UsageType' column not found in long format.")

else:
    print("\n--- Skipping analysis on long data: Reshaping failed or resulted in an empty DataFrame ---")

print("\n--- Script Finished ---")
```
